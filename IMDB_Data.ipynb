```python
import requests
import pandas as pd
import time
```

```python
# Load the CSV file
netflix_data = pd.read_csv('netflix_titles 2.csv')
```

```python
netflix_data.shape
```

```python
netflix_data_sample=netflix_data.head(8807)
```

```python
# Define the OMDb API key and base URL
omdb_api_key = '3ca24c50'
base_url = 'http://www.omdbapi.com/'
```

```python
# Function to query the OMDb API
def get_omdb_details(title, year):
    
    params = {
        't': title,
        'y': year,
        'apikey': omdb_api_key
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        print(response.status_code)
        return response.json()  # Return the JSON response
    return None
```

```python
# Fetch OMDb details for each title with appropriate delay
omdb_details_list = []
for index, row in netflix_data_sample.iterrows():
    title = row['title']
    year = row['release_year']
    omdb_details = get_omdb_details(title, year)
    if omdb_details and omdb_details.get('Response') == 'True':
        omdb_details_list.append(omdb_details)
    else:
        omdb_details_list.append({})
    time.sleep(1)  # Add a 1-second delay between requests
```

```python
# Convert the list of OMDb details to a DataFrame
omdb_details_df = pd.DataFrame(omdb_details_list)
```
